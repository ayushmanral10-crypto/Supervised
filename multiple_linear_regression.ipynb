## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 12)
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

## Training the Multiple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() # model name
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = regressor.predict(X_test)   # y_test should not be passed as it is the ground truth
np.set_printoptions(precision=1)   # 1 decimal place only
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))  # converting 1D to 2D array
```

## Evaluating the Model Performance

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
r2_score(y_test, y_pred)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))  # rmse- root mean square error
```

```python
mean_absolute_error(y_test, y_pred)   # np.abs(y_test - y_pred)
```
